```python
import pandas as pd
import re

df = pd.read_csv('dataset.csv')
```

```python
df['selftext'] = df['selftext'].fillna('')
df['selftext'] = df['selftext'].replace({'[removed]': '', '[deleted]': ''}, regex=False)
df['full_text'] = (df['title'] + ' ' + df['selftext']).str.strip()
```

```python
def clean_for_bert(text):
    text = str(text)
    text = re.sub(r'http\S+', '', text)     
    text = re.sub(r'\s+', ' ', text).strip() 
    return text
```

```python
df['bert_text'] = df['full_text'].apply(clean_for_bert)
df[['label', 'bert_text']].to_csv('dataset_bert.csv', index=False)
```
