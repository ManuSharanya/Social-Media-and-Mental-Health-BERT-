```python
from sklearn.model_selection import train_test_split
import pandas as pd

df=pd.read_csv('dataset_bert.csv')
```

```python
train_bert, test_bert = train_test_split(
    df[['label', 'bert_text']],
    test_size=0.2,
    stratify=df['label'],
    random_state=42   
)
```

```python
train_bert.to_csv('train_bert.csv', index=False)
test_bert.to_csv('test_bert.csv', index=False)
```
